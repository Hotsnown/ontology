```python
import ast
import numpy as np

def fiches_arrêt(index): 
    if type(rupture.iloc[index]["zones"]) is str:
      zones = ast.literal_eval(rupture.iloc[index]["zones"])
      introduction = extract_section(index, zones, "introduction")
      expose = extract_section(index, zones, "expose")
      premier_moyen = extract_moyen(0)
      deuxième_moyen = extract_moyen(1)
      troisième_moyen = extract_moyen(2)
      quatrième_moyen = extract_moyen(3)
      motivations = extract_section(index, zones, "motivations")
      dispositif = extract_section(index, zones, "dispositif")
      annexes = extract_section(index, zones, "annexes")
      print(zones.get("moyens"))
      return [introduction, expose, moyens, motivations, dispositif, annexes]
    else:
      return np.full(6, np.nan)

def extract_section(index, zones, header):
   try:
    start = zones.get(header)[0]["start"]
    end =  zones.get(header)[0]['end']
    return rupture.iloc[index]["text"][start:end]
   except:
      return np.nan
   
def extract_moyen(index, zones, numéro):
  try:
    start = zones.get("moyens")[numéro]["start"]
    end =  zones.get("moyens")[numéro]['end']
    return rupture.iloc[index]["text"][start:end]
   except:
      return np.nan
   
fiches = []
for i in range(len(rupture)):
  ret = fiches_arrêt(i)
  fiches.append(ret)

structure = pd.DataFrame(fiches, columns=["introduction", "expose", "moyens", "motivations", "dispositif", "annexes"])
structure
```
